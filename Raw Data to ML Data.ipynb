# 2. Data Adjustment

## Adjustment the Data Variables in each Features

```python
# Upgrade the data for suitable in ML study
# DIABETE4 = Diabetes disease
# 1 is for yes diabetes
# Change the variable of 2 (only during pregnancy), 3 (no diabetes) and 4 (pre-diabetes or borderline diabetes) to 0
# Remove the variables of 7 (dont knows) and 9 (refused)
brfss['DIABETE4'] = brfss['DIABETE4'].replace({2:0, 3:0, 4:0})
brfss = brfss[brfss.DIABETE4 != 7]
brfss = brfss[brfss.DIABETE4 != 9]
# Display the variable in DIABETE4 after upgrade
brfss.DIABETE4.unique()
```

```python
# Upgrade the data for suitable in ML study
# CHCKDNY2 = Kidney disease
# 1 is for yes kidney diseases
# Change the variable of 2 (no kidney disease) to 0
# Remove the variables of 7 (dont knows) and 9 (refused)
brfss['CHCKDNY2'] = brfss['CHCKDNY2'].replace({2:0})
brfss = brfss[brfss.CHCKDNY2 != 7]
brfss = brfss[brfss.CHCKDNY2 != 9]
# Display the variable in CHCKDNY2 after upgrade
brfss.CHCKDNY2.unique()
```

```python
# Upgrade the data for suitable in ML study
# ASTHMA3 = Asthma disease
# 1 is for yes asthma diseases
# Change the variable of 2 (no asthma disease) to 0
brfss['ASTHMA3'] = brfss['ASTHMA3'].replace({2:0})
brfss = brfss[brfss.ASTHMA3 != 7]
brfss = brfss[brfss.ASTHMA3 != 9]
# Display the variable in ASTHMA3 after upgrade
brfss.ASTHMA3.unique()
```

```python
# Upgrade the data for suitable in ML study
# CVDCRHD4 = Heart disease
# 1 is for yes heart diseases
# Change the variable of 2 (no heart disease) to 0
# Remove the variables of 7 (dont knows) and 9 (refused)
brfss['CVDCRHD4'] = brfss['CVDCRHD4'].replace({2:0})
brfss = brfss[brfss.CVDCRHD4 != 7]
brfss = brfss[brfss.CVDCRHD4 != 9]
# Display the variable in CVDCRHD4 after upgrade
brfss.CVDCRHD4.unique()
```

```python
# Upgrade the data for suitable in ML study
# CVDSTRK3 = Stroke
# 1 is for yes stroke
# Change the variable of 2 (no stroke) to 0
# Remove the variables of 7 (dont knows) and 9 (refused)
brfss['CVDSTRK3'] = brfss['CVDSTRK3'].replace({2:0})
brfss = brfss[brfss.CVDSTRK3 != 7]
brfss = brfss[brfss.CVDSTRK3 != 9]
# Display the variable in CVDSTRK3 after upgrade
brfss.CVDSTRK3.unique()
```

```python
# Upgrade the data for suitable in ML study
# _MICHD = Heart attack
# 1 is for yes heart attack
# Change the variable of 2 (no heart attack) to 0
brfss['_MICHD'] = brfss['_MICHD'].replace({2: 0})
# Display the variable in _MICHD after upgrade
brfss._MICHD.unique()
```

```python
# Upgrade the data for suitable in ML study
# CHCSCNCR = Skin cancer
# 1 is for yes skin cancer
# Change the variable of 2 (no skin cancer) to 0
# Remove the variables of 7 (dont knows) and 9 (refused)
brfss['CHCSCNCR'] = brfss['CHCSCNCR'].replace({2:0})
brfss = brfss[brfss.CHCSCNCR != 7]
brfss = brfss[brfss.CHCSCNCR != 9]
# Display the variable in CHCSCNCR after upgrade
brfss.CHCSCNCR.unique()
```
